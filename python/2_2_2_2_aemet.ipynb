```python
import untangle
import matplotlib.pyplot as plt
%matplotlib inline

# analizar el fichero
obj = untangle.parse('http://www.aemet.es/xml/municipios/localidad_28079.xml')

l = obj.root.prediccion.dia

# extracción de datos
fechas = []
maximas = []
minimas = []

for dia in l:
    fechas.append(dia['fecha'][8:])
    maximas.append(int(dia.temperatura.maxima.cdata))
    minimas.append(int(dia.temperatura.minima.cdata))

# gráfica
plt.plot(fechas,maximas,label="máximas")
plt.plot(fechas,minimas,label="mínimas")
plt.title("Fechas desde el "+fechas[0]+" hasta el "+fechas[-1]+" del mes en curso")
plt.legend()
plt.show()
```
